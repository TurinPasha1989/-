Обьем произвдства

```python
prod = pd.read_excel('Market_data.xlsx', 'производство').T
prod.head()
```

```python
prod = prod.set_index(prod[0])
prod.head()
```

```python
prod.index.names = ['Date']
prod.head()
```

```python
prod = prod.drop(prod.columns[[0,1,2,12,13,14,15]], axis='columns')
prod.head()
```

```python
prod = prod.loc[prod.index.notnull()]
prod.head()
```

```python
prod.rename(columns={3: 'Российская Федерация',
                  4: 'Центральный федеральный округ',
                  5: 'Северо-Западный федеральный округ',
                  6: 'Южный федеральный округ (с 2010 года)',
                  7: 'Северо-Кавказский федеральный округ',
                  8: 'Приволжский федеральный округ',
                  9: 'Уральский федеральный округ',
                  10: 'Сибирский федеральный округ',
                  11: 'Дальневосточный федеральный округ'}, inplace=True)
prod.head(10)
```

```python
prod.info()
```

```python
prod.plot(figsize=(16,9))
```

Пропущенные значения заменяю средними

```python
prod = prod.fillna(prod.mean())
```

```python
prod.info()
```

```python
prod.plot(figsize=(16,9))
```

Производство практически не имеет тренда к росту, округа, где производство занимает большую долю имеют сезонность.

Доли производства по округам

```python
vals = [prod['Центральный федеральный округ'].sum(), 
        prod['Северо-Западный федеральный округ'].sum(), 
        prod['Южный федеральный округ (с 2010 года)'].sum(), 
        prod['Северо-Кавказский федеральный округ'].sum(),
        prod['Приволжский федеральный округ'].sum(),
        prod['Уральский федеральный округ'].sum(),
        prod['Сибирский федеральный округ'].sum(),
        prod['Дальневосточный федеральный округ'].sum(),
       ]
labels = ['Центральный федеральный округ', 'Северо-Западный федеральный округ', 
          'Южный федеральный округ (с 2010 года)', 'Северо-Кавказский федеральный округ',
          'Приволжский федеральный округ', 'Уральский федеральный округ', 
          'Сибирский федеральный округ', 'Дальневосточный федеральный округ'
         ]

plt.pylab.pie(vals, labels=labels)
```
